# ***Question 1:***
##### Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
# connecting to the webdriver:
driver = webdriver.Chrome(r"C:\Users\Lenovo\chromedriver.exe")

# Opening 'Amazon.in' chrome browser:
url='http://www.amazon.in/'
driver.get(url)
driver.maximize_window()
time.sleep(3)

# Taking input from user about product search:
User_input=input('Enter the title of Product you are interest in search :')
```

# ***Question 2:***
##### In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Making Empty list to scrap data into them:

Brand =[]
Product = []
Price =[]
Return =[]
Delivery =[]
Availability =[]
```

```python
# Scraping “Product URL” of all products listed in first 3 pages: 
# //a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]
URL= []
# range(0,3) used to scrape three pages on website
for page in range(0,3):
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        URL.append(i.get_attribute('href'))
        time.sleep(2)

# locating next page button and clicking
Nxt_page = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
Nxt_page.click()

URL
```

```python
len(URL)
```

```python
for i in URL:
    driver.get(i)
    time.sleep(2)
    
    # Extracting "Brand Name":
    try:
        brand = driver.find_element(By.XPATH,'//a[@id="bylineInfo"]')
        Brand.append(brand.text) 
    except NoSuchElementException:
        Brand.append('-')
        time.sleep(1)
        
    # Extracting "Name of the Product":
    try:
        product = driver.find_element(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
        Product.append(product.text)
    except NoSuchElementException:
        Product.append('-')
    time.sleep(1)
    
    # Extracting "Price": 
    try:
        price = driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
    time.sleep(1)
    
    # Extracting "Return/Exchange":
    try:
        returns = driver.find_element(By.XPATH,'//a[@class="a-size-small a-link-normal a-text-normal"]')
        Return.append(returns.text)
    except NoSuchElementException:
        Return.append('-')
    time.sleep(1)
    
    # Extracting "Expected Delivery":
    try:
        delivery = driver.find_element(By.XPATH,'//span[@class="a-text-bold"]')
        Delivery.append(delivery.text)
    except NoSuchElementException:
        Delivery.append('-')
    time.sleep(1)
    
    # Extracting "Availability":
    try:
        Avail = driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')
        Availability.append(Avail.text)
    except NoSuchElementException:
        Availability.append('-')
    time.sleep(1)
```

```python
len(Brand),len(Product),len(Price),len(Return),len(Delivery),len(Availability)
```

```python
# Creating DataFrame:
Search_Item = pd.DataFrame()
Search_Item['Product_URL'] = URL
Search_Item['Brand Name'] = Brand
Search_Item['Name of the Product'] = Product
Search_Item['Price '] = Price 
Search_Item['Return/Exchange'] = Return
Search_Item['Expected Delivery'] = Delivery
Search_Item['Availability'] = Availability

# naming the index:
Search_Item.index.name = "S.No." 

# Index Numbering in Ascending order:
Search_Item.index += 1

# printing the DataFrame:
Search_Item
```

```python
# Closing of Automated Window:

driver.close()
```

# ***Question 10:***
##### Write a python program to scrape data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description.

```python
# Connecting to webdriver
driver = webdriver.Chrome(r"C:\Users\Lenovo\chromedriver.exe")
driver.maximize_window()
time.sleep(1)

# Opening url in chrome browser
url="https://www.hostelworld.com"
driver.get(url)
time.sleep(2)
```

```python
driver.find_element(By.XPATH,'//*[@id="location-text-input-field"]').click()
```

```python
#locating the location search bar
Search = driver.find_element(By.XPATH,'//*[@id="search-input-field"]')
Search.send_keys("London")
time.sleep(1)
```

```python
#select london
london = driver.find_element(By.XPATH,'//*[@id="predicted-search-results"]/li[2]/div')
london.click()
time.sleep(1)
```

```python
# clicking on "Let's go!" button:
Search_button = driver.find_element(By.XPATH,'//*[@id="search-button"]')
Search_button.click()
```

```python
#fetching hostel name:

hostel_name = []

try:
    name = driver.find_elements(By.XPATH,"//h2[@class='title title-6']")
    for i in name:
        hostel_name.append(i.text)
            
except NoSuchElementException:
    hostel_name.append('-')

hostel_name
```

```python
#fetching distance from city centre

distance = []

try:
    dist = driver.find_elements(By.XPATH,"//span[@class='description']")
    for i in dist:
        distance.append(i.text.split(' ')[-4])
except NoSuchElementException:
    distance.append('-')
    
distance
```

```python
# fetching ratings

rating = []

try:
    rat = driver.find_elements(By.XPATH,"//div[@class='rating rating-summary-container big']")
    for i in rat:
        rating.append(i.text.split('\n')[0])
except NoSuchElementException:
    rating.append('-')

rating
```

```python
#fetching total reviews

reviews = [] 

try:
    rws = driver.find_elements(By.XPATH,"//div[@class='reviews']")
    for i in rws:
        reviews.append(i.text.split(' ')[0])
except NoSuchElementException:
    reviews.append('-')

reviews
```

```python
#fetch overall review

over_all = [] 

try:
    overall_rw = driver.find_elements(By.XPATH,"//div[@class='keyword']")
    for i in overall_rw:
        over_all.append(i.text)
except NoSuchElementException:
    over_all.append('-')
over_all = over_all[2:]
over_all
```

```python
# fetching Privates from price:

pvt_prices=[]
pvt_price = driver.find_elements(By.XPATH,"//div[@class='price-col']")
for i in pvt_price:
    pvt_prices.append(i.text.replace('Rs30021.62 ',' ').split('\n')[-1]) 

# Using list slicing getting odd indexed elements:
pvt_prices = pvt_prices[::2] # pvt_prices[1::2] ->(for even)
pvt_prices
```

```python
# fetching Dorms from price:

dorms_prices = []
dorms_price = driver.find_elements(By.XPATH,"//a[@class='prices']")
for i in dorms_price:
    dorms_prices.append(i.text.split('\n')[-1])
dorms_prices = dorms_prices[2:]

# split each element in list and keep second part
dorms_prices = [item.split(' ', 1)[-1].replace('Dorms Available','-') for item in dorms_prices]
dorms_prices
```

#### Extracting "Property Description" :

```python
# lets fetch url of each hostel:

product_url = []

p_url = driver.find_elements(By.XPATH,"//a[@class='prices']")
for i in p_url:    
    product_url.append(i.get_attribute('href'))

product_url = product_url[2:]
product_url 
```

```python
# Extracting 'facilities':
facilities = []
property_description =[] 

for i in product_url:
    driver.get(i)

    # fetch 'facilities':
    try:
        facility = driver.find_elements(By.XPATH,"//ul[@class='facility-sections']")
        for i in facility:
            facilities.append(i.text)
    except NoSuchElementException:
        facilities.append('-')
   
    # Extracting 'Product Description':     
    try:
        pd = driver.find_element(By.XPATH,'/html/body/div[3]/div/div/div[2]/section/div[6]/div/div[2]/div/div/div[2]')
        property_description.append(pd.text)
    except NoSuchElementException:
        property_description.append("No Description")
```

```python
property_description
```

```python
facilities
```

```python
# Creating dataframe:
Hostel_World = pd.DataFrame({})
Hostel_World['Hostel Name'] = hostel_name
Hostel_World['Distance from City Centre'] = distance
Hostel_World['Ratings'] = rating
Hostel_World['Total Reviews'] = reviews
Hostel_World['Overall Reviews'] = over_all
Hostel_World['Privates from price'] = pvt_prices
Hostel_World['Dorms from price'] = dorms_prices
Hostel_World['Facilities'] = facilities
Hostel_World['Property Description'] = property_description
Hostel_World["Hostel URL's"] = product_url
Hostel_World
```
